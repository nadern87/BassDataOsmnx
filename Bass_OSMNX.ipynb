```python
## create smart function
tag = {}
def get_pois_via_tag(center_point, tag, dist= 5000):
    output =ox.geometries_from_point(center_point, tag, dist)
    pd.DataFrame(output)
    dc_ = []
    if not output.empty:
        count = len(output) #we need the count of rows, because every row represent of of the tags
        dc_.append(count)
    else: 
        count = 0
        dc_.append(count)
    return dc_
```

```python
#Road types
# define function to get the roads data for each coordinates and return a dictionary or road types from column(highway)

def get_col_dict(center_point, tag, col):
    d = ox.geometries_from_point(center_point, tags = tag, dist= 1000)
    pd.DataFrame(d)
    count = d[col].value_counts()
    dc = count.to_dict()
    return dc
```

```python
# define the Key, and which column of data you want to return as dictionary
import time

tag = {'highway': True}
col = "highway"

roads_list=[]
for index, row in df.iterrows():
    tmp = get_col_dict(row["Point"], tag, col)
    roads_list.append(tmp)
    print(index)
    time.sleep(25)

print('Process is done!!')
```

```python
# Get Road Lanes

tag = {'lanes': True}
def get_lanes(center_point):
    gdf = ox.geometries_from_point(center_point, tag, dist=1000)
    if gdf.empty:
        return 0
    else:
        count = gdf.lanes.astype(int).sum()
        return count
```

```python
lanes_list = df['Point'].apply(lambda p: get_lanes(p)).tolist()
```

```python
lanes_df = pd.DataFrame.from_dict(lanes_list)
lanes_df.columns = ['aggr_lanes']
lanes_df # todo combine this
```

```python
roadnetwork_list = [lanes_df,roards_df]
roadnetwork_df = pd.concat(roadnetwork_list, axis=1)
roadnetwork_df.to_csv('bass_road_dataset_1000m.csv')
```

```python
# Get POIS

ids= df['Id'].tolist()
points = df['Point'].tolist()

def get_pois(ids, points):
    pois = {}
    for i, point in zip(ids, points):
        gdf = ox.geometries_from_point(point, tags={'amenity':True}, dist=1000)
        if gdf.empty:
            pois[i] = {}
        else:
            df = pd.DataFrame(gdf)
            counts = df['amenity'].value_counts().to_dict()
            pois[i] = counts
    return pois
```

```python
pois = get_pois(ids, points)
print(pois)
```

```python
amenities_df = pd.DataFrame.from_dict(pois)
bass_amenities = amenities_df.T
bass_amenities
```

```python
# Replace NaN with zeros and check again
bass_amenities = bass_amenities.fillna(0)
bass_amenities.isnull().sum().sum()
```

```python
bass_amenities.columns
```

```python
edu_list= ['school', 'library']
bass_amenities['aggr_education'] = bass_amenities[edu_list].sum(axis=1)
bass_amenities['aggr_education'].astype(int)
food_list = ['fast_food', 'bar', 'cafe', 'restaurant']
bass_amenities['aggr_food'] = bass_amenities[food_list].sum(axis=1)
bass_amenities['aggr_food'].astype(int)
vehicle_list = ['parking','fuel','car_wash']
bass_amenities['aggr_automobile'] = bass_amenities[vehicle_list].sum(axis=1)
bass_amenities['aggr_automobile'].astype(int)
money_list = ['bank']
bass_amenities['aggr_money'] = bass_amenities[money_list].sum(axis=1)
religion_list = ['place_of_worship'] # need to add the landuse=cemetary
bass_amenities['aggr_religion'] = bass_amenities[religion_list].sum(axis=1) 
medical_list = ['hospital', 'pharmacy', 'dentist', 'nursing_home']
bass_amenities['aggr_health'] = bass_amenities[medical_list].sum(axis=1)
entertainment_list = ['cinema', 'arts_centre']
bass_amenities['aggr_entertainment'] = bass_amenities[entertainment_list].sum(axis=1)
public_service_list= ['fire_station', 'post_office', 'prison', 'police', 'community_centre', 'townhall', 'courthouse']
bass_amenities['aggr_public_service'] = bass_amenities[entertainment_list].sum(axis=1)
```
